### Trend analysis 

- here is a 
[list of tests](https://www.bae.ncsu.edu/programs/extension/wqg/319monitoring/TechNotes/technote6_trend_analysis.pdf) 
for trend analysis and 
[the description of MK test](http://vsp.pnnl.gov/help/Vsample/Design_Trend_Mann_Kendall.htm)

- for the python implementation of the MK test, credits to 
[Sat Kumar Tomer](http://pydoc.net/Python/ambhas/0.4.0/ambhas.stats/)

Preparing dataframe for tests:

```python
median_age_df = full_df[full_df.Race.isin(popular_races.keys())].groupby(['Race','RaceYear'],as_index=False).agg({'Year': np.median})
median_age_df.rename(inplace=True,columns={'Year': 'median_age'})
median_age_df.head()
```

Running test on __median__ age:

```python
import ambhas_stats
```

```python
print('Significance threshold for MK test = 0.05')

for my_race in median_age_df.Race.unique():
    
    trend,h,p,z = ambhas_stats.mk_test(np.asarray(median_age_df.median_age[median_age_df.Race==my_race]),0.05) 
    print('p-value for',my_race,round(p,3),'- trend:',trend,'- significant?',h)
```

__Conclusions on mean age across time__

For many of the races analyzed is _not possible to claim a global trend_ for the runners' mean age, across time.      
Note that the test is affected by the _lack of data points in the time series_ (due to time reasons, we did not correct for this issue). 

However, the trends flagged as significant (p<0.05) by the MK test seem reasonable:

- an _increase_ for ASICS Bremgarter Reusslauf,Jungfrau-Marathon,Kerzerslauf
- a _decrease_ for instance 20km de Lausanne, Course de l'Escalade, Hallwilerseelauf, Luzerner Stadtlauf, Morat-Fribourg, Thuner Stadtlauf, Zürcher Silvesterlauf

### Fitting p-l for runners per race 

preparing the data:

```python
data_for_fit  = full_df.Race.value_counts().values
```

perform the fit:

```python
fit_runners_race = powerlaw.Fit(data_for_fit,discrete=True)

print('fitted exponent:',fit_runners_race.alpha)
print('fitting error:',fit_runners_race.sigma)

print('xmin of the fit:',fit_runners_race.xmin)
print('xmax of the fit:',fit_runners_race.xmax)
```

plot the results - mind that here we use _logarithmically spaced bins_:

```python
powerlaw.plot_pdf(data_for_fit,marker='.',linestyle='',label='data')
fit_runners_race.power_law.plot_pdf(label='p-l fit')

plt.legend()
plt.ylabel('number or races',size=16)
plt.xlabel('number or runners/race',size=16);
```

# Statistics on Performance VS Weather

One of the main study on this topic, from 
[2007](http://journals.lww.com/acsm-msse/pages/articleviewer.aspx?year=2007&issue=03000&article=00012&type=abstract), found correlation between 
[WBGT](https://en.wikipedia.org/wiki/Wet-bulb_globe_temperature)
and decrease of performance, both for male and female runners - 
we don't have WBGT data, but we could try to compute it following this 
[paper](https://pdfs.semanticscholar.org/c43f/78bf1cc042130280f8aaa6b6725c1df49c75.pdf) or this
[website](http://www.climatechip.org/heat-stress-index-calculation).       
(An older study from 
[1997](http://journals.lww.com/acsm-msse/pages/articleviewer.aspx?year=1997&issue=09000&article=00018&type=abstract)
